# Comparison of Complexity

```python
from time import time
from math import log, factorial
import random
import pandas as pd
import matplotlib.pyplot as plt

random.seed(42)
plt.style.use('dark_background')
```

```python
# functions of time complexity
t_func = {
    "O(1)": lambda n: 1,
    "O(log(n))": lambda n: log(n),
    "O(n)": lambda n: n,
    "O(n log(n))": lambda n: n * log(n),
    "O(n^2)": lambda n: n * n,
    "O(2^n)": lambda n: 2 ** n,
    "O(n!)": lambda n: factorial(n),
}
```

```python
# plot data
N = range(1, 100 + 1)
t = [[t_func[k](n) for k in t_func] for n in N]

fig, ax = plt.subplots(figsize=[8, 6], dpi=100)
ax.set_title("Big-O Complexity")
ax.set_ylabel("Operations")
ax.set_xlabel("Input Size")
ax.set_xlim([1, 100])
ax.set_ylim([0, 1300])
ax.plot(t)
ax.legend(t_func.keys());
fig.savefig("complexity.png", dpi=120, bbox_inches="tight")
```

```python
def formatter(k, n):
    """Convert numbers to string."""
    number = t_func[k](n)
    s = str(number)
    if "." in s:
        return f"$${number:0.3f}$$"
    if len(s) < 6:
        return f"$${s}$$"
    return f"$${s[0]}.{s[1:3]}\times10^{{{len(s)-1}}}$$"


# table data
N = [8, 16, 32, 64, 128, 256]
t = [[formatter(k, n) for k in t_func] for n in N]
df = pd.DataFrame(t, columns=t_func.keys(), index=N)
df
```
